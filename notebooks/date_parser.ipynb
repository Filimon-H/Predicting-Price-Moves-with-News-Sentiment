```python


import pandas as pd
from pathlib import Path
from dateutil import parser
# Path to the CSV file (from inside the notebooks/ folder)
DATA_DIR = Path("..") / "data"
CSV_FILE = DATA_DIR / "raw_analyst_ratings.csv"   # change file name if needed

df = pd.read_csv(CSV_FILE)
```

```python
def parse_date_safe(x):
    x = str(x).strip()
    
    # 1. Try automatic universal parser (handles timezone + ISO)
    try:
        return parser.parse(x)
    except:
        pass
    
    # 2. Try explicit US format M/D/YYYY H:M
    try:
        return pd.to_datetime(x, format="%m/%d/%Y %H:%M")
    except:
        pass
    
    # 3. Try explicit US format MM/DD/YYYY H:M
    try:
        return pd.to_datetime(x, format="%m/%d/%Y %H:%M")
    except:
        pass

    # 4. Try US format without time
    try:
        return pd.to_datetime(x, format="%m/%d/%Y")
    except:
        pass
    
    # If everything fails
    return pd.NaT
```

```python
df['date'] = df['date'].apply(parse_date_safe)
```

```python
if pd.api.types.is_datetime64tz_dtype(df['date']):
    df['date'] = df['date'].dt.tz_localize(None)
```

```python
# Ensure the 'date' column is converted to datetimes and made timezone-naive
df['date'] = pd.to_datetime(df['date'], errors='coerce', utc=True)
if pd.api.types.is_datetime64tz_dtype(df['date']):
	df['date'] = df['date'].dt.tz_localize(None)

print("Start date:", df['date'].min())
print("End date:", df['date'].max())
```
